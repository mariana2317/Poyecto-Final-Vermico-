```python
import pandas as pd
df_temp = pd.read_csv("Datos_temperatura (1) (1).csv", encoding="latin1", sep=";")
df_temp.columns = df_temp.columns.str.strip()  
```

```python
df_temp["mes"] = ((df_temp["dia"] - 1) // 30) + 1
# Promedio mensual
df_temp["promedio mensual temperatura"] = df_temp.groupby("mes")["temperatura C째"].transform("mean")

# Promedio anual
df_temp["promedio anual temperatura"] = df_temp["temperatura C째"].mean()
```

```python
df_hum = pd.read_csv("promedio final humedad.csv", sep=";")
df_hum.columns = df_hum.columns.str.strip()
df_hum["humedad_promedio"] = df_hum[["humedad 1", "humedad 2"]].mean(axis=1)
```

```python
df_hum["mes"] = (df_hum["dias"] // 30) + 1
# Promedio mensual
df_hum["promedio mensual humedad"] = df_hum.groupby("mes")["humedad_promedio"].transform("mean")

# Promedio anual
df_hum["promedio anual humedad"] = df_hum["humedad_promedio"].mean()
```
